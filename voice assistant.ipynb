```python
import speech_recognition as sr
from groq import Groq
from gtts import gTTS
import pygame
import tempfile


groq_client = Groq(api_key="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")
history = []

def listen():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        recognizer.adjust_for_ambient_noise(source, duration=0.5)
        
        
        try:
            audio = recognizer.listen(source, timeout=5)
        except sr.WaitTimeoutError:
            
            return None

    
    try:
        text = recognizer.recognize_google(audio)
        print(f"You said: {text}")
        return text
    except sr.UnknownValueError:
        
        return None
    except sr.RequestError as e:
        # API was unreachable or unresponsive
        print(f"Could not request results; {e}")
        return None


def ask_llm(user_text):
    history.append({"role": "user", "content": user_text})
    response = groq_client.chat.completions.create(
       model="llama-3.1-8b-instant",   
        messages=[{"role": "system", "content": "You are a helpful voice assistant. Keep answers short and conversational."}] + history
    )
    reply = response.choices[0].message.content
    history.append({"role": "assistant", "content": reply})
    return reply


def speak(text):
    tts = gTTS(text=text, lang="en")
    path = "response.mp3"
    tts.save(path)
    pygame.mixer.init()
    pygame.mixer.music.load(path)
    pygame.mixer.music.play()
    while pygame.mixer.music.get_busy():
        pygame.time.Clock().tick(10)
    pygame.mixer.music.unload()


def main():
    print("Voice Assistant ready. Say 'Exit' to exit.")
    while True:
        user_input = listen()
        if not user_input:
            continue
        if "Exit" in user_input.lower():
            speak("Goodbye!")
            break
        reply = ask_llm(user_input)
        print(f"Assistant: {reply}")
        speak(reply)

if __name__ == "__main__":
    main()
```
